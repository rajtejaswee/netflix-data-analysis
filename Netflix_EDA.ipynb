## Handling the missing values

```python
# Fill missing values with 'Unknown' for categorical fields
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
# Drop rows where date_added is null (only 10 rows)
df.dropna(subset=['date_added'], inplace=True)
```

```python
# Fill rating/duration with placeholder
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
#Strip leading/trailing spaces
df['date_added'] = df['date_added'].str.strip()
```

```python
#Converting letting pandas infer mixed formats
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
#Extracting the year, month and day.
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['day_of_week'] = df['date_added'].dt.day_name()
```

```python
# Reseting the index
df.reset_index(drop=True, inplace=True)
```

# Explanatory Analysis

### Movies vs Tv Shows on Neflix

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Netflix color palette
netflix_red = '#db0000'
netflix_dark = '#000000'
netflix_gray = '#564d4d'
netflix_blood = '#831010'
white = '#ffffff'

# Data
type_counts = df['type'].value_counts()
percentages = (type_counts / type_counts.sum()) * 100

# Plot
plt.figure(figsize=(8, 3), facecolor=white)
bars = plt.barh(type_counts.index, type_counts.values, color=netflix_red, edgecolor=netflix_blood, height=0.4)

# Add percentage labels next to bars
for i, (value, pct) in enumerate(zip(type_counts.values, percentages)):
    plt.text(value + 50, i, f'{pct:.1f}%', va='center', color=netflix_dark, fontsize=13, fontweight='bold')

# Styling
plt.title('Netflix Content Breakdown: Movies vs TV Shows', fontsize=18, color=netflix_dark, weight='bold', pad=15)
plt.xlabel('Number of Titles', fontsize=12, color=netflix_dark, labelpad=10)
plt.yticks(color=netflix_dark, fontsize=12)
plt.xticks(color=netflix_dark)
plt.gca().set_facecolor(white)
plt.grid(axis='x', color=netflix_gray, linestyle='--', linewidth=0.5)
plt.box(False)

plt.tight_layout()
plt.show()
```

### Content Added per Year

```python
import matplotlib.pyplot as plt

# Filter out invalid or old years
year_counts = df['year_added'].value_counts().sort_index()
year_df = year_counts.reset_index()
year_df.columns = ['year', 'count']
year_df = year_df[year_df['year'] >= 2008]  # Netflix mostly has data after 2008

# Sort by year
year_df = year_df.sort_values('year')

# Smart height management
bar_height = 0.4
base_height = len(year_df) * bar_height + 2
fig_height = min(base_height, 15)  # Keep under 15 inches

# Plotting
plt.figure(figsize=(10, fig_height), facecolor=white)
colors = [netflix_red if i % 2 == 0 else netflix_gray for i in range(len(year_df))]
bars = plt.barh(year_df['year'].astype(str), year_df['count'], color=colors, edgecolor=netflix_dark)

# Add labels
for i, (value, label) in enumerate(zip(year_df['count'], year_df['year'])):
    plt.text(value + 5, i, str(value), va='center', color=netflix_dark, fontsize=11, fontweight='bold')

# Styling
plt.title('Content Added to Netflix by Year', fontsize=18, color=netflix_dark, weight='bold', pad=15)
plt.xlabel('Number of Titles', fontsize=12, color=netflix_dark)
plt.xticks(color=netflix_dark)
plt.yticks(color=netflix_dark, fontsize=11)
plt.grid(axis='x', color=netflix_gray, linestyle='--', linewidth=0.5)
plt.gca().set_facecolor(white)
plt.box(True)

plt.tight_layout()
plt.show()



```

### Top 10 Countries by Content

```python
import matplotlib.pyplot as plt

# Step 1: Handle multiple countries per row
country_series = df['country'].dropna().str.split(', ')
country_exploded = country_series.explode()

# Step 2: Count top 10 countries
top_countries = country_exploded.value_counts().head(10)

# Step 3: Prepare the plot
fig, ax = plt.subplots(figsize=(10, 6), facecolor=white)

# Step 4: Horizontal bars
bars = ax.barh(
    range(len(top_countries)),                      # Positions
    top_countries.values[::-1],                     # Values (reversed for highest on top)
    color= netflix_red,
    edgecolor=netflix_dark
)

# Step 5: Add labels to the right of the bars
for i, value in enumerate(top_countries.values[::-1]):
    ax.text(value + 5, i, str(value), va='center', color=netflix_dark, fontsize=11, fontweight='bold')

# Step 6: Styling
ax.set_yticks(range(len(top_countries)))  # Explicitly set y-ticks
ax.set_yticklabels(top_countries.index[::-1], color=netflix_dark, fontsize=11)

ax.set_title("Top 10 Countries with Most Netflix Titles", fontsize=18, color=netflix_dark, weight='bold', pad=15)
ax.set_xlabel("Number of Titles", fontsize=12, color=netflix_dark)
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticks(), color=netflix_dark)
ax.grid(axis='x', color=netflix_gray, linestyle='--', linewidth=0.5)
ax.set_facecolor(white)
ax.tick_params(left=False)
plt.box(False)
plt.tight_layout()
plt.show()

```

```python
# Group and pivot the data
rating_type = df.groupby(['type', 'rating']).size().unstack().fillna(0)

# Transpose to get Ratings on x-axis and Type stacked
ax = rating_type.T.plot(kind='bar', stacked=True, figsize=(12, 7),
                        color=[netflix_red, netflix_gray], edgecolor=netflix_blood)

# Titles and labels
plt.title('Distribution of Ratings by Content Type', fontsize=16, color=netflix_dark, weight='bold', pad=15)
plt.xlabel('Rating', fontsize=12, color=netflix_dark)
plt.ylabel('Number of Titles', fontsize=12, color=netflix_dark)

# Ticks
plt.xticks(rotation=45, color=netflix_dark, fontsize=11)
plt.yticks(color=netflix_dark)

# Styling
ax.set_facecolor(white)
plt.grid(axis='y', linestyle='--', linewidth=0.5, color=netflix_gray)
plt.tight_layout()
plt.legend(title='Type', title_fontsize=12, fontsize=11)

plt.show()

```

### Genres (from 'listed_in' column)

```python
from collections import Counter

# Step 1: Extract all genres
all_genres = df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in all_genres for genre in sublist]

# Step 2: Count occurrences
genre_counts = Counter(flat_genres)
top_genres = genre_counts.most_common(10)

# Step 3: Prepare data
genres, counts = zip(*top_genres)
colors = [netflix_red if i % 2 == 0 else netflix_gray for i in range(len(genres))]
```

```python
from collections import Counter

# Step 1: Split and flatten genre values
all_genres = df['listed_in'].dropna().str.split(', ')
flat_genres = [genre for sublist in all_genres for genre in sublist]

# Step 2: Count genre frequencies
genre_counts = Counter(flat_genres)
top_genres = genre_counts.most_common(10)

# Step 3: Prepare data
genres, counts = zip(*top_genres)
colors = [netflix_red if i % 2 == 0 else netflix_gray for i in range(len(genres))]

# Step 4: Plotting
fig, ax = plt.subplots(figsize=(10, 6), facecolor=white)
bars = ax.barh(genres[::-1], counts[::-1], color=colors[::-1], edgecolor=netflix_blood)

# Add value labels
for i, v in enumerate(counts[::-1]):
    ax.text(v + 2, i, str(v), va='center', color=netflix_dark, fontsize=11, fontweight='bold')

# Style
ax.set_title('Top 10 Most Common Netflix Genres', fontsize=16, color=netflix_dark, weight='bold', pad=15)
ax.set_xlabel('Number of Titles', fontsize=12, color=netflix_dark)
ax.tick_params(axis='x', colors=netflix_dark)
ax.tick_params(axis='y', colors=netflix_dark)
ax.grid(axis='x', linestyle='--', linewidth=0.5, color=netflix_gray)
ax.set_facecolor(white)
fig.patch.set_facecolor(white)

plt.tight_layout()
plt.show()
```

### When is Netflix most active in uploads?

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['month_year'] = df['date_added'].dt.to_period('M')
monthly_counts = df['month_year'].value_counts().sort_index()
monthly_df = monthly_counts.reset_index()
monthly_df.columns = ['Month', 'Count']
```

```python
monthly_df['Month'] = monthly_df['Month'].dt.strftime('%Y-%m')  # Convert to string format like '2021-07'

plt.figure(figsize=(16, 6), facecolor=white)
plt.bar(monthly_df['Month'], monthly_df['Count'], color=netflix_red, edgecolor=netflix_blood)

# Styling
plt.title('Netflix Content Uploads Over Time (Monthly)', fontsize=18, color=netflix_dark, weight='bold')
plt.xlabel('Month', fontsize=12, color=netflix_dark)
plt.ylabel('Number of Titles Added', fontsize=12, color=netflix_dark)
plt.xticks(rotation=90, fontsize=9, color=netflix_dark)
plt.yticks(color=netflix_dark)
plt.grid(axis='y', linestyle='--', linewidth=0.5, color=netflix_gray)
plt.tight_layout()
plt.gca().set_facecolor(white)

plt.show()
```
