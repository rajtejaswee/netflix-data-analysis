## Handling the missing values

```python
# Fill missing values with 'Unknown' for categorical fields
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
# Drop rows where date_added is null (only 10 rows)
df.dropna(subset=['date_added'], inplace=True)
```

```python
# Fill rating/duration with placeholder
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
```

```python
#Strip leading/trailing spaces
df['date_added'] = df['date_added'].str.strip()
```

```python
#Converting letting pandas infer mixed formats
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
#Extracting the year, month and day.
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['day_of_week'] = df['date_added'].dt.day_name()
```

```python
# Reseting the index
df.reset_index(drop=True, inplace=True)
```

# Explanatory Analysis

### Movies vs Tv Shows on Neflix

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Netflix color palette
netflix_red = '#db0000'
netflix_dark = '#000000'
netflix_gray = '#564d4d'
netflix_blood = '#831010'
white = '#ffffff'

# Data
type_counts = df['type'].value_counts()
percentages = (type_counts / type_counts.sum()) * 100

# Plot
plt.figure(figsize=(8, 3), facecolor=white)
bars = plt.barh(type_counts.index, type_counts.values, color=netflix_red, edgecolor=netflix_blood, height=0.4)

# Add percentage labels next to bars
for i, (value, pct) in enumerate(zip(type_counts.values, percentages)):
    plt.text(value + 50, i, f'{pct:.1f}%', va='center', color=netflix_dark, fontsize=13, fontweight='bold')

# Styling
plt.title('Netflix Content Breakdown: Movies vs TV Shows', fontsize=18, color=netflix_dark, weight='bold', pad=15)
plt.xlabel('Number of Titles', fontsize=12, color=netflix_dark, labelpad=10)
plt.yticks(color=netflix_dark, fontsize=12)
plt.xticks(color=netflix_dark)
plt.gca().set_facecolor(white)
plt.grid(axis='x', color=netflix_gray, linestyle='--', linewidth=0.5)
plt.box(False)

plt.tight_layout()
plt.show()
```

### Content Added per Year

```python
import matplotlib.pyplot as plt

# Filter out invalid or old years
year_counts = df['year_added'].value_counts().sort_index()
year_df = year_counts.reset_index()
year_df.columns = ['year', 'count']
year_df = year_df[year_df['year'] >= 2008]  # Netflix mostly has data after 2008

# Sort by year
year_df = year_df.sort_values('year')

# Smart height management
bar_height = 0.4
base_height = len(year_df) * bar_height + 2
fig_height = min(base_height, 15)  # Keep under 15 inches

# Plotting
plt.figure(figsize=(10, fig_height), facecolor=white)
colors = [netflix_red if i % 2 == 0 else netflix_gray for i in range(len(year_df))]
bars = plt.barh(year_df['year'].astype(str), year_df['count'], color=colors, edgecolor=netflix_dark)

# Add labels
for i, (value, label) in enumerate(zip(year_df['count'], year_df['year'])):
    plt.text(value + 5, i, str(value), va='center', color=netflix_dark, fontsize=11, fontweight='bold')

# Styling
plt.title('Content Added to Netflix by Year', fontsize=18, color=netflix_dark, weight='bold', pad=15)
plt.xlabel('Number of Titles', fontsize=12, color=netflix_dark)
plt.xticks(color=netflix_dark)
plt.yticks(color=netflix_dark, fontsize=11)
plt.grid(axis='x', color=netflix_gray, linestyle='--', linewidth=0.5)
plt.gca().set_facecolor(white)
plt.box(True)

plt.tight_layout()
plt.show()



```

### Top 10 Countries by Content

```python
import matplotlib.pyplot as plt

# Step 1: Handle multiple countries per row
country_series = df['country'].dropna().str.split(', ')
country_exploded = country_series.explode()

# Step 2: Count top 10 countries
top_countries = country_exploded.value_counts().head(10)

# Step 3: Prepare the plot
fig, ax = plt.subplots(figsize=(10, 6), facecolor=white)

# Step 4: Horizontal bars
bars = ax.barh(
    range(len(top_countries)),                      # Positions
    top_countries.values[::-1],                     # Values (reversed for highest on top)
    color= netflix_red,
    edgecolor=netflix_dark
)

# Step 5: Add labels to the right of the bars
for i, value in enumerate(top_countries.values[::-1]):
    ax.text(value + 5, i, str(value), va='center', color=netflix_dark, fontsize=11, fontweight='bold')

# Step 6: Styling
ax.set_yticks(range(len(top_countries)))  # Explicitly set y-ticks
ax.set_yticklabels(top_countries.index[::-1], color=netflix_dark, fontsize=11)

ax.set_title("Top 10 Countries with Most Netflix Titles", fontsize=18, color=netflix_dark, weight='bold', pad=15)
ax.set_xlabel("Number of Titles", fontsize=12, color=netflix_dark)
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticks(), color=netflix_dark)
ax.grid(axis='x', color=netflix_gray, linestyle='--', linewidth=0.5)
ax.set_facecolor(white)
ax.tick_params(left=False)
plt.box(False)
plt.tight_layout()
plt.show()

```

### When is Netflix most active in uploads?

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['month_year'] = df['date_added'].dt.to_period('M')
monthly_counts = df['month_year'].value_counts().sort_index()
monthly_df = monthly_counts.reset_index()
monthly_df.columns = ['Month', 'Count']
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Make sure 'monthly_df' exists with columns 'Month' and 'Count'
monthly_df['Month'] = monthly_df['Month'].dt.to_timestamp()
monthly_df['Quarter'] = monthly_df['Month'].dt.to_period('Q')
quarterly_df = monthly_df.groupby('Quarter')['Count'].sum().reset_index()
quarterly_df['Quarter'] = quarterly_df['Quarter'].astype(str)

# Define Netflix colors
netflix_red = '#db0000'
netflix_gray = '#564d4d'
netflix_dark = '#000000'
netflix_blood = '#831010'
white = '#ffffff'

# Plot
fig, ax1 = plt.subplots(figsize=(14, 6), facecolor=white)
ax1.bar(quarterly_df['Quarter'], quarterly_df['Count'], color=netflix_gray)
ax1.set_ylabel('Number of Titles Added', color=netflix_dark)
ax1.set_xlabel('Quarter', color=netflix_dark)
ax1.set_title('Netflix Upload Trend (Quarterly)', fontsize=16, weight='bold', color=netflix_dark)
ax1.tick_params(axis='x', rotation=45)
ax1.tick_params(axis='y', labelcolor=netflix_dark)
ax1.set_facecolor(white)

# Line overlay
ax2 = ax1.twinx()
ax2.plot(quarterly_df['Quarter'], quarterly_df['Count'], color=netflix_red, marker='o', linewidth=2)
ax2.tick_params(axis='y', labelcolor=netflix_red)
ax2.set_ylabel('Trend', color=netflix_red)

ax1.grid(axis='y', linestyle='--', linewidth=0.5, color=netflix_gray)
fig.tight_layout()
plt.show()
```

```python
#Genre Popularity Over Time on Netflix

import pandas as pd
import matplotlib.pyplot as plt

# Colors
netflix_red = '#db0000'
netflix_dark = '#000000'
netflix_white = '#ffffff'
netflix_gray = '#564d4d'

# Prepare year column
df['date_added'] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year

# Drop rows with missing data
genre_df = df.dropna(subset=['listed_in', 'year_added'])

# Explode genres
genre_df['listed_in'] = genre_df['listed_in'].str.split(', ')
exploded_genre_df = genre_df.explode('listed_in')

# Group by year and genre
genre_trend = exploded_genre_df.groupby(['year_added', 'listed_in']).size().unstack(fill_value=0)

# Select top 5-7 genres based on total appearances
top_genres = genre_trend.sum().sort_values(ascending=False).head(7).index
genre_trend_top = genre_trend[top_genres]

# Plot
plt.figure(figsize=(12, 6), facecolor=netflix_white)
for genre in genre_trend_top.columns:
    plt.plot(genre_trend_top.index, genre_trend_top[genre], label=genre, linewidth=2.5)

plt.title('Genre Popularity Over Time on Netflix', fontsize=16, color=netflix_dark)
plt.xlabel('Year', fontsize=12, color=netflix_dark)
plt.ylabel('Number of Titles', fontsize=12, color=netflix_dark)
plt.legend(title='Genre', frameon=False)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Clean and prepare date
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.dropna(subset=['date_added'])

# Extract year and month
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month_name()

# Order months correctly
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
df['month_added'] = pd.Categorical(df['month_added'], categories=month_order, ordered=True)

# Group by year and month
heatmap_data = df.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)

# Plot
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, cmap='Reds', linewidths=0.3, linecolor='gray', annot=True, fmt='d')

plt.title('Netflix Content Uploads by Month and Year', fontsize=16, color='#db0000')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Year', fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Filter the data from 2014 onwards
heatmap_trimmed = heatmap_data[heatmap_data.index >= 2014]

# Plot the heatmap
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_trimmed, cmap='Reds', annot=True, fmt='d')

plt.title('Netflix Content Uploads by Month and Year (2014 onwards)', fontsize=16, color='#db0000')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Year', fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Word Cloud of Content Descriptions
from wordcloud import WordCloud, STOPWORDS

# Combine all descriptions into one string
text = ' '.join(df['description'].dropna().astype(str).values)

# Define stopwords
stopwords = set(STOPWORDS)

# Generate the word cloud
wordcloud = WordCloud(
    stopwords=stopwords,
    background_color='white',
    max_words=200,
    width=1600,
    height=800,
    colormap='Reds'
).generate(text)

# Plotting the word cloud
plt.figure(figsize=(16, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Netflix Content Descriptions', fontsize=18, color=netflix_red)
plt.tight_layout()
plt.show()
```
